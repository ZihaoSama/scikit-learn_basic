```python
#ID:300595912
#Name:Zihao Zhang
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt  
from sklearn.datasets import make_blobs,make_classification,make_circles

ds1,la1= make_blobs(n_samples=1000,n_features=2,random_state=42)
ds2,la2= make_classification(n_samples=1000,n_features=2,n_clusters_per_class=1,n_informative=2,random_state=42,n_redundant=0)
ds3,la3= make_circles(n_samples=1000,noise=0.3,random_state=42)
Ds1=[ds1,la1]
Ds2=[ds2,la2]
Ds3=[ds3,la3]
#set dictionary which is used to "for" loop
toySet_dic = {
    "make_blobs":Ds1,
    "make_classification":Ds2,
    "make_circles":Ds3
}
```

(a) K-Means

```python
# For KMeans
from sklearn.cluster import KMeans
# plot settings
flg = plt.figure(figsize=(15,5))
flg = plt.figure(1)
#subpot pos setting
pos=1
for name,datasets in toySet_dic.items():
    #ValueError: setting an array element with a sequence. 
    #The requested array has an inhomogeneous shape after 2 dimensions.
    #The detected shape was (2, 1000) + inhomogeneous part.
    toyData=datasets[0]
    #model setting
    kmeans_Model = KMeans(random_state=42,n_clusters=3).fit(toyData)
    #subplot drawing
    plt.subplot(1,3,pos)
    pos+=1
    plt.title(name + " on KMEAM")
    #(x,y,color of sequence of colors for markers)
    plt.scatter(toyData[:,0], toyData[:,1], c=kmeans_Model.labels_)
plt.show()
```
